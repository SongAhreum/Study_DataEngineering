차이



- 설명

df2 = df:

df2는 df의 참조를 유지합니다. 즉, df2를 변경하면 df도 함께 변경됩니다. 이는 두 변수가 같은 DataFrame 객체를 가리키기 때문입니다.
df2 = df.dropna(inplace=True):

dropna(inplace=True)는 결측값이 제거된 DataFrame을 반환하지 않고, 원래 DataFrame (df)을 직접 변경합니다. 이 경우 df2는 None을 가리키게 됩니다.
따라서 df2는 원래 DataFrame인 df가 아닌 None을 가리키게 됩니다.

### 사용자정의함수 dataFrame에 적용시키기


```python
#함수
def add_cm(height):
  return str(height)+"cm"
```

```python
#사용법
df['키']=df['키'].apply(add_cm)
df
```

```python
def capitalize(lang):
  if pd.notnull(lang) :
    return lang.capitalize()
```

```python
# df['SW특기'] = df['SW특기'].capitalize()  error::::AttributeError: 'Series' object has no attribute 'capitalize'
df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
