```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

```python
df = pd.read_csv('./data/PolynomialRegressionData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
X,y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5),dpi=30)
plt.scatter(X,y,color='blue')
plt.plot(X,reg.predict(X),color='green')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
reg.score(X,y)
```

# 정리

```python
df = pd.read_csv('./data/PolynomialRegressionData.csv')
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python

poly_reg = PolynomialFeatures(degree=2) #2차 방정식 객체 생성
X_poly = poly_reg.fit_transform(X)
```

```python
reg = LinearRegression()
reg.fit(X_poly,y)
```

```python
X_range = np.arange(min(X), max(X), 0.1)
```

```python
X_range=X_range.reshape(-1, 1)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X_range, reg.predict(poly_reg.fit_transform(X_range)), color='green')
plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
pred=reg.predict(poly_reg.fit_transform([[4.5]]))
pred
```
